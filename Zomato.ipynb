# **8. Analyze the distribution of restaurants based on features like Wi-Fi, Alcohol availability, etc. Investigate if the presence of certain features correlates with higher ratings.**


```python
df['has_wifi'] = df['highlights'].fillna('').str.contains('Free Wif')
df['has_alcohol'] = df['highlights'].fillna('').str.contains('Alcohol', case=False)
```

```python
df[["has_wifi","has_alcohol"]]
```

```python
wifi_rating = df.groupby('has_wifi')['aggregate_rating'].mean().reset_index()
alcohol_rating = df.groupby('has_alcohol')['aggregate_rating'].mean().reset_index()
```

```python
wifi_rating.round(1)
```

```python
wifi_aval=df.groupby("has_wifi")["name"].count().reset_index()
wifi_aval.sort_values(by="name",ascending=False)
```

```python
alcohol_aval=df.groupby("has_alcohol")["name"].count().reset_index()
alcohol_aval.sort_values(by="name",ascending=False)
```

```python
plt.pie(wifi_aval['name'],labels=wifi_aval['has_wifi'],autopct='%1.1f%%')
plt.title('Resturants that have availablity of Wifi')
plt.tight_layout()
plt.show()
```

```python
plt.pie(alcohol_aval['name'],labels=alcohol_aval['has_alcohol'],autopct='%1.1f%%')
plt.title('Resturants that have availablity of alcohol')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
sns.barplot(data=wifi_rating, x='has_wifi', y='aggregate_rating', palette='Blues')
plt.title("Average Rating by Wi-Fi Availability")
plt.xticks([0, 1], ['No Wi-Fi', 'Wi-Fi'])
plt.ylabel("Average Rating")

plt.subplot(1, 2, 2)
sns.barplot(data=alcohol_rating, x='has_alcohol', y='aggregate_rating', palette='Reds')
plt.title("Average Rating by Alcohol Availability")
plt.xticks([0, 1], ['No Alcohol', 'Alcohol'])
plt.ylabel("Average Rating")

plt.tight_layout()
plt.show()
```
